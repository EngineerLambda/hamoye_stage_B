#### Question 12
<b>From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two d.p.?

```python
x = file[["T2"]]
y = file.T6
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 42)
model_t = LinearRegression()
model_t.fit(x_train,y_train)
```

```python
pred_t = model_t.predict(x_test)
r2 = r2_score(y_test,pred_t)
round(r2,2)
```

#### Question 13
<b>Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. The target variable is “Appliances”. Answer the following questions:

What is the Mean Absolute Error (in two decimal places)?

```python
file = file.drop(columns = ["date","lights"])
file.head()
```

```python
scaler = MinMaxScaler()
scaled = scaler.fit_transform(file)
norm = pd.DataFrame(scaled, columns = file.columns)
norm.head()
```

```python
model_m = LinearRegression()
x_n = norm.drop(columns = ["Appliances"])
y_n = norm["Appliances"]
nx_train, nx_test, ny_train, ny_test = train_test_split(x_n,y_n, test_size = 0.3, random_state = 42)

model_m.fit(nx_train,ny_train)
pred_m = model_m.predict(nx_test)
```

```python
mae = mean_absolute_error(ny_test,pred_m)
round(mae,2)
```

#### Question 17
<b> Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?

```python
def get_weight_df(model, feat, col_name):
    weights = pd.Series(model.coef_,feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ["Features",col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
model_m_weight = get_weight_df(model_m,nx_train,"Weights")
model_m_weight
```
